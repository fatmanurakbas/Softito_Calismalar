### 🔸 Soru 1:
Aşağıdaki değişkenleri tanımlayın ve her birinin tipini `type()` fonksiyonu ile yazdırın:
- `sehir`: "Ankara" (string)

### 🔸 Soru 5:
`cumle = "  python programlama dili çok güçlüdür  "` değişkeni için:
- Baştaki ve sondaki boşlukları temizleyin (`.strip()`)
- İlk harfini büyük yapın (`.capitalize()`)

```python
# Cevabınızı buraya yazın:
cumle = " python programlama dili çok güçlüdür "

temiz = cumle.strip()
buyuk = temiz.capitalize()
kelime_sayisi = len(temiz.split())

print(temiz)
print(buyuk)
print(kelime_sayisi)
print("güçlü" in temiz)
```

### 🔸 Soru 8:

### 🔸 Soru 10:
`while` döngüsü ile kullanıcının girdiği sayıya (n=5) kadar Fibonacci dizisini yazdırın.

Örnek çıktı: `0, 1, 1, 2, 3`

```python
# Cevabınızı buraya yazın:
n = 5

a = 0
b = 1
i = 0

while i < n:
    print(a, end=", " if i < n - 1 else "")
    a, b = b, a + b
    i += 1
```

### 🔸 Soru 12:
Aşağıdaki listede:
- En büyük ve en küçük elemanı bulun
- Listeyi sıralayın
- 7 sayısının listede kaç kez geçtiğini bulun
- Listeyi tersine çevirin

```python
# Cevabınızı buraya yazın:
liste = [4, 7, 2, 7, 9, 1, 5]

print("En küçük:", min(liste))
print("En büyük:", max(liste))

liste.sort()
print("Sıralı liste:", liste)

print("7 sayısı adedi:", liste.count(7))

liste.reverse()
print("Ters liste:", liste)
```

### 🔸 Soru 13 (List Comprehension):
List comprehension kullanarak:
- 1-20 arasındaki çift sayıların listesini oluşturun
- Bir kelime listesindeki 4'ten uzun kelimeleri filtreleyin

```python
# Cevabınızı buraya yazın:
cift_sayilar = [x for x in range(1, 21) if x % 2 == 0]
print(cift_sayilar)

kelimeler = ["elma", "armut", "çilek", "muz", "karpuz", "ananas"]
uzun_kelimeler = [k for k in kelimeler if len(k) > 4]
print(uzun_kelimeler)
```

### 🔸 Soru 14:
Bir kitap sözlüğü oluşturun (`baslik`, `yazar`, `yil`, `sayfa_sayisi`) ve:
- Tüm anahtar-değer çiftlerini döngüyle yazdırın

```python
# Cevabınızı buraya yazın:
kitap = {
    "baslik": "Suç ve Ceza",
    "yazar": "Dostoyevski",
    "yil": 1866,
    "sayfa_sayisi": 500
}

for anahtar, deger in kitap.items():
    print(anahtar, ":", deger)

kitap["fiyat"] = 120

del kitap["yazar"]

print(kitap)
```

### 🔸 Soru 15:
Bir sayının asal olup olmadığını kontrol eden `asal_mi(n)` fonksiyonu yazın.
- 1'den 20'ye kadar olan asal sayıları bu fonksiyonu kullanarak yazdırın.

```python
# Cevabınızı buraya yazın:
def asal_mi(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

for i in range(1, 21):
    if asal_mi(i):
        print(i)
```

### 🔸 Soru 16:
Bir listenin ortalamasını, en büyük ve en küçük elemanını döndüren `liste_istatistik(liste)` fonksiyonu yazın.
Fonksiyon 3 değer döndürsün: `(ortalama, minimum, maksimum)`

```python
# Cevabınızı buraya yazın:
def liste_istatistik(liste):
    ortalama = sum(liste) / len(liste)
    minimum = min(liste)
    maksimum = max(liste)
    return ortalama, minimum, maksimum

sayilar = [4, 7, 2, 9, 1, 5]

ort, min_deger, max_deger = liste_istatistik(sayilar)

print("Ortalama:", ort)
print("Minimum:", min_deger)
print("Maksimum:", max_deger)
```

### 🔸 Soru 18 (Bütünleştirici):
Bir **not defteri uygulaması** yazın:
- `notlar` adında boş bir liste oluşturun
- `not_ekle(metin)` fonksiyonu: Notu listeye ekler
- `not_listele()` fonksiyonu: Tüm notları numaralandırarak yazdırır
- `not_sil(index)` fonksiyonu: Verilen indexteki notu siler, hatalı index için hata mesajı verir

Fonksiyonları test edin: 3 not ekleyin, listeleyin, ortadakini silin, tekrar listeleyin.

```python
# Cevabınızı buraya yazın:
notlar = []

def not_ekle(metin):
    notlar.append(metin)

def not_listele():
    for i, n in enumerate(notlar):
        print(i, n)

def not_sil(index):
    try:
        notlar.pop(index)
    except IndexError:
        print("Hatalı index!")

not_ekle("Matematik çalış")
not_ekle("Python tekrar et")
not_ekle("Proje yap")

not_listele()
not_sil(1)
not_listele()
```
